```python
# Question 1.1
# before trying to run any of this please ensure you have installed all modules from requirements.txt

%matplotlib widget 
import matplotlib.pyplot as plt
from data_utils import load_dataset
from nearest import knn, shuffle           # Nearest is a handmade module that abstracts KNN for regression 
import numpy as np
np.random.seed(0)                 # We set the seed so the psedonumber generator produces the same randomness
from scipy.interpolate import interp1d # This is just used for plotting

def l1(diff):
    return np.sum(abs(diff), axis=1)
    
    
```

## Mauna_loa Data set

```python
# Here we test the Mauna_loa data set with
#l2 distance
xtrain, xvalid, xtest, ytrain, yvalid, ytest = load_dataset('mauna_loa')#, n_train=1000, d=2)
X = np.vstack([xtrain, xvalid]) # create a combined X since we are doing Cross Validation
Y = np.vstack([ ytrain,yvalid ])# create a combined Y 
shuffle(X, Y)

scores = []
predicted = []
folds = 100   # the range of ks we would like to consider, has nothing to do with folds
for k in range(1, folds):
    nn = knn(k=k)
    scores.append(nn.batch_cv_knn(X,Y, 5)) # we run batch cross validation with X,Y and 5 folds
    predicted.append(nn.pred)              # predictions from cross validation are brought out

fig = plt.figure()
plt.plot( [j for j in range(1, folds)], scores)
plt.xlabel("k")
plt.ylabel("rmse")
plt.title("RMSE compared to Knn")
plt.show()

[print(f'{j + 1}: rmse: {scores[j]}') for j in range(folds - 1)]; # The remaining sections are similar
```

```python
#l1 distance
xtrain, xvalid, xtest, ytrain, yvalid, ytest = load_dataset('mauna_loa')#, n_train=1000, d=2)
X = np.vstack([xtrain, xvalid])
Y = np.vstack([ ytrain,yvalid ])
shuffle(X, Y)

scores = []
predicted = []
folds = 100
for k in range(1, folds):
    nn = knn(k=k, dist_type = "l1")
    scores.append(nn.batch_cv_knn(X,Y, 5))
    predicted.append(nn.pred)

fig = plt.figure()
plt.plot( [j for j in range(1, folds)], scores)
plt.xlabel("k")
plt.ylabel("rmse")
plt.title("RMSE compared to Knn")
plt.show()

[print(f'{j + 1}: rmse: {scores[j]}') for j in range(folds - 1)];
```

```python
# Test RMSE
xtrain, xvalid, xtest, ytrain, yvalid, ytest = load_dataset('mauna_loa')
X = np.vstack([xtrain, xvalid])                                                              # -me search as before
Y = np.vstack([ ytrain,yvalid ])
shuffle(X, Y)

nn = knn(k=2, dist_type='l1')    # we have found our estimated K and prefered distance metric from above
nn.test_error(X, Y,xtest, ytest) # we use it here to get the test RMSE
```

```python
# For Mauna Loa
# We see that the Minimum RMSE Occurs when k = 2
# We also notice that the distance metric doesn't have much of an effect here
# this makes sense as this is a one dimentional data set in features.
# If all features are collinear, (or on the same line due to dimensionality) the
# notion of "distance" is the same regardless since there is only one feature

# As per the instructions we plot the predicted against the true points
# You can zoom in to see the difference between k values

def plot(X, Y, marker="-"):
    true_line_visx = np.linspace(X.min(), X.max(), 500)
    true_line_visy = interp1d(X[:,0], Y[:,0], kind='quadratic')(true_line_visx)
    plt.plot(true_line_visx,true_line_visy, marker,  markersize=1)
    

fig = plt.figure()
plot(X,predicted[1])
plot(X,predicted[8])

plot(X,predicted[6])
plot(X,predicted[4])
plot(X,predicted[12])
plot(X,predicted[14])
plot(X,Y)


plt.legend(["k = 2","k = 9","k = 7","k = 5","k = 13","k = 15", "True"])
plt.xlabel("x")
plt.ylabel("Predicted Values")
plt.title("Predicted Values")
plt.show()
```

```python

# Here we plot forecasted values as per request from Q1
# all of them are pretty bad

test_preds ={}
for k in [5,8,11,14,17]:
    nn = knn(k=k, dist_type='l2')
    test_preds[k] = nn.batch_re(np.vstack([xvalid, xtrain ]), np.vstack([yvalid, ytrain ]), xtest)


fig = plt.figure()
for k in test_preds:
    plt.plot(xtest, test_preds[k], "-", markersize=1)
plt.plot(xtest, ytest, "-", markersize=1)

plt.legend([str(keys) for keys in test_preds] + ["True"])
plt.xlabel("x")
plt.ylabel("Predicted Values")
plt.title("Predicted Values")
plt.show()
```

```python
# l2
xtrain, xvalid, xtest, ytrain, yvalid, ytest = load_dataset('rosenbrock', n_train=1000, d=2) # we engage in the sa-
X = np.vstack([xtrain, xvalid])                                                              # -me search as before
Y = np.vstack([ ytrain,yvalid ])
shuffle(X, Y)

scores = []
predicted = []
folds = 100
for k in range(1, folds):
    nn = knn(k=k, dist_type = "l2")
    scores.append(nn.batch_cv_knn(X,Y, 5))
    predicted.append(nn.pred)

fig = plt.figure()
plt.plot( [j for j in range(1, folds)], scores)
plt.xlabel("k")
plt.ylabel("rmse")
plt.title("RMSE compared to Knn")
plt.show()

[print(f'{j + 1}: rmse: {scores[j]}') for j in range(folds - 1)];

r_score = np.mean(scores)
```
